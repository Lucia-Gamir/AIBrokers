```python
import pandas as pd
from transformers import pipeline
```

```python
sentiment_analyzer = pipeline("sentiment-analysis")
train_data = pd.read_json("finer_train_data.json")
```

```python
train_data
```

```python
sentences_train = list(train_data[0])
results_train = sentiment_analyzer(sentences_train)
```

```python
sentiments = [i["label"] for i in results_train]
```

```python
for i in range(len(train_data)):
    train_data.iloc[i, 1]["sentiment"] = sentiments[i]
```

```python
import pandas as pd
import json

# Supongamos que ya tienes cargado tu DataFrame
# df = pd.read_json("tu_archivo.json", lines=True)

# Convertimos el DataFrame en una lista con el formato deseado
dataset = []
for _, row in train_data.iterrows():
    text = row[0]
    annotations = row[1]
    
    # Aseguramos que el campo 'sentiment' exista
    if "sentiment" not in annotations:
        annotations["sentiment"] = "neutral"  # o cualquier valor por defecto
    
    dataset.append([text, annotations])

# Guardamos en un archivo JSON
with open("train_labels.json", "w", encoding="utf-8") as f:
    json.dump(dataset, f, indent=4, ensure_ascii=False)
```

```python
from collections import Counter

# Extraemos todos los valores de sentimiento
sentiments = [row[1].get("sentiment", "undefined") for row in dataset]

# Contamos cada tipo
counter = Counter(sentiments)

# Mostramos los resultados
print("Distribución de clases de sentimiento:")
for sentiment, count in counter.items():
    print(f"{sentiment}: {count}")
```

```python
# Display the results in a nice format
for sentence, result in zip(sentences_train, results_train):
    label = result['label']
    score = result['score']
    print(f"Sentence: '{sentence}'")
    print(f"Sentiment: {label} (confidence: {score:.2f})")
    print('-' * 60)
```
